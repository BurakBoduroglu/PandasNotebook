## Series

```python
import numpy as np
import pandas as pd
```

```python
myDict = {"Apple" : 100, "Peach" : 120, "Banana" : 110}
```

```python
pd.Series(myDict)
```

```python
numberList = [100, 120, 110]
nameList = ["Apple", "Peach", "Banana"]
```

```python
pd.Series(numberList, nameList )
```

```python
pd.Series(data = numberList, index = nameList)
```

```python
pd.Series(["Name1","Name2","Name3"], [1,2,3])
```

```python
compResult1 = pd.Series(data = [10, 5, 3], index = ["Name1", "Name2", "Name3"])
```

```python
compResult1
```

```python
compResult2 = pd.Series(data = [20, 12, 9], index = ["Name1", "Name2", "Name3"])
```

```python
compResult2
```

```python
compResult2["Name1"]
```

```python
lastResult = compResult2 + compResult1
```
